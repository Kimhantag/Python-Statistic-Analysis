```python
import numpy as np
import pandas as pd
from scipy.stats import chisquare
from scipy.stats import chi2_contingency

desa=pd.read_csv("대사중후군.csv", encoding="CP949")
desa.head()
```

```python
label1={1:'실험군A', 2:'실험군B', 3:'대조군'}
desa['group']=desa['group'].map(label1)
label2={1:'건강매우좋음', 2:'건강좋음', 3:'보통', 4:'건강나쁨', 5:'건강매우나쁨'}
desa['건강']=desa['건강'].map(label2)
print(desa.group, desa.건강)
```

```python
pd.crosstab(desa.group, desa.건강, margins=True, margins_name='전체')
```

```python
pd.crosstab(desa.group, desa.건강, margins=True, margins_name='전체',
           normalize='all').round(2)
```

```python
table=pd.crosstab(desa.group, desa.건강)
chi, p, df, expected=chi2_contingency(table)
expected_table=pd.DataFrame(data=expected, index=table.index,
                           columns=table.columns)
expected_table
```

```python
print("chisquare=%.3f"%chi, "P-value=%.3f"%p)
```

```python
선호도=np.tile(['찬성', '무관심', '반대'], 2)
count=np.array([138, 83, 64, 64, 67, 84])
정당=np.repeat(['민주당', '공화당'], 3)
data={'선호도':선호도, '정당':정당, 'count':count}
```

```python
table1=pd.crosstab(index=data['정당'], columns=data['선호도'], 
                   values=data['count'], aggfunc=sum)
chi, p, df, expected=chi2_contingency(table1)
expected_table1=pd.DataFrame(data=expected, index=table1.index,
                           columns=table1.columns)
expected_table1
```
