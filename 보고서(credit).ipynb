```python
#한글폰트 파일실행
!python han-font.py
exec(open('han-font.py').read())
import numpy as np
import pandas as pd

credit=pd.read_csv("credit.csv", encoding="CP949")
credit.head()
```

```python
credit1=pd.get_dummies(credit)
credit1
```

```python
print('credit 데이터셋 크기: ', credit.shape)
credit.info()
```

```python
# 질적변수를 더미변수로 변환
# 질적변수 -> 더미변수
credit_data1 = pd.get_dummies(credit.loc[:,['checking_balance', 'credit_history', 'savings_balance',
                                            'employment_duration', 'housing', 'job']])
# 양적변수 & target
credit_data2=credit.loc[:, ['months_loan_duration', 'amount', 'percent_of_income',
                            'years_at_residence', 'age', 'existing_loans_count', 'default']]
# 분석에 사용되는 데이터셋
credit_data=credit_data1.join(credit_data2)
credit_data.head()
```

```python
# 독립변수(수치형 자료)와 종속변수(target)
# 독립변수 6개 이름과 회귀계수 값을 매칭시켜 출력(zip - 매칭하여 묶어줌)
x_data=credit_data.iloc[:,:-1]
y_data=credit_data["default"]
```

```python
#훈련용 데이터와 평가용 데이터 분할하기
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size = 0.2, random_state = 0)
```

```python
# x_train과 x_test를 StandardScaler를 이용해 정규화
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
x_train=pd.DataFrame(x_train, columns=x_data.columns)
x_test=pd.DataFrame(x_test, columns=x_data.columns)
x_train.head()
```

```python
#1. 로지스틱 회귀
#로지스틱 회귀 분석: (1) 모델 생성
from sklearn.linear_model import LogisticRegression
credit_LR = LogisticRegression()
```

```python
#로지스틱 회귀 분석: (2) 모델 훈련 및 모형적합결과
reg=credit_LR.fit(x_train, y_train)
```

```python
# 로지스틱 회귀 분류 정확도
print( "score =", reg.score(x_test, y_test))
```

```python
# 독립변수 이름과 회귀계수 값을 매칭시켜 출력(zip - 매칭하여 묶어줌)
for f, w in zip(x_train.columns, np.exp(reg.coef_[0])):
  print(f"{f:<25} {w}")
```

```python
#로지스틱 회귀 분석: (3) 평가자료에 대한 예측 수행 -> 예측 결과 Y_predict 구하기 
from sklearn.metrics import confusion_matrix, accuracy_score 
from sklearn.metrics import precision_score, recall_score
y_pred = credit_LR.predict(x_test)
print(confusion_matrix(y_test, y_pred))
```

```python
# 정확도, 정밀도, 재현성 -> 정확도 중요
acccuracy = accuracy_score(y_test, y_pred) 
precision = precision_score(y_test, y_pred, pos_label="yes") 
recall = recall_score(y_test, y_pred, pos_label="yes") 
print('정확도: {0:.3f}, 정밀도: {1:.3f}, 재현율: {2:.3f}'.format(acccuracy,precision,recall))
# 모형평가(분류정확도_보고서) 종합정리
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
# svm - 의미있는 양적자료 사용( ) 
# 'months_loan_duration', 'amount', 'percent_of_income', 'age', 'existing_loans_count
numeric_xtrain=x_train.loc[:, ['months_loan_duration', 'amount', 'percent_of_income',
                              'age', 'existing_loans_count']]
numeric_xtest=x_test.loc[:, ['months_loan_duration', 'amount', 'percent_of_income',
                              'age', 'existing_loans_count']]
numeric_xtrain.head()
```
