```python
### 이항-로지스틱회귀모형분석
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer

b_cancer = load_breast_cancer()
b_cancer     #dictionary{  }형태의 자료
```

```python
# 독립변수(b_cancer.feature)로 DataFrame 자료형으로 변환
b_cancer_df = pd.DataFrame(b_cancer.data, columns = b_cancer.feature_names)
b_cancer_df['diagnosis']= b_cancer.target  
```

```python
# diagnosis는 악성이면 1, 양성이면 0의 값이므로 유방암 여부에 대한 이진 분류
b_cancer_df.head()
```

```python
print('유방암 진단 데이터셋 크기: ', b_cancer_df.shape)
b_cancer_df.info()
```

```python
# 정규분포의 표준화(scaler=standardScaler())함수를 scaler함수로 생성
from sklearn.preprocessing import StandardScaler 
scaler= StandardScaler()
b_cancer_s = scaler.fit_transform(b_cancer.data)

# data[0] ~ 첫번째 행
print(b_cancer.data[0]) #첫번째 변수의 원본데이터
print(b_cancer_s[0])    #첫번쨰 변수의 표준화 데이터
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import train_test_split

# input=X, target=Y 설정하기 
Y = b_cancer_df['diagnosis'] 
X = b_cancer_s
```

```python
#훈련용 데이터와 평가용 데이터 분할하기 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)

#로지스틱 회귀 분석: (1) 모델 생성 
LR_b_cancer = LogisticRegression()
```

```python
#로지스틱 회귀 분석: (2) 모델 훈련 및 모형적합결과
reg=LR_b_cancer.fit(X_train, Y_train)
print(reg.coef_)
print(np.exp(reg.coef_))
print( "R²=", reg.score(X_train, Y_train))
```

```python
# 독립변수 30개 이름과 회귀계수 값을 매칭시켜 출력(zip - 매칭하여 묶어줌)
# np.exp(reg.coef_[0])
for f, w in zip(b_cancer.feature_names, np.exp(reg.coef_[0])):
  print(f"{f:<25} {w}")
```

```python
#로지스틱 회귀 분석: (3) 훈련자료와 평가자료에 대한 예측 수행 -> 예측 결과 Y_predict 구하기 
Y_fit=LR_b_cancer.predict(X_train)
Y_pred = LR_b_cancer.predict(X_test)
```

```python
#오차 행렬( )
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_train, Y_fit)) 
print(confusion_matrix(Y_test, Y_pred))
```

```python
# 정확도, 정밀도, 재현성
from sklearn.metrics import accuracy_score, precision_score, recall_score
acccuracy = accuracy_score(Y_test, Y_pred) 
precision = precision_score(Y_test, Y_pred) 
recall = recall_score(Y_test, Y_pred) 
print('정확도: {0:.3f}, 정밀도: {1:.3f}, 재현율: {2:.3f}'.format(acccuracy,precision,recall))
```

```python
# 모형평가(분류정확도_보고서) 종합정리
from sklearn.metrics import classification_report
print(classification_report(Y_test, Y_pred))
```

```python
### 다항-로지스틱회귀모형분석 : iris 
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
# input=X, target=Y 설정하기 
X = iris.data
Y = iris.target
```

```python
# iris_df 데이터프레임에 독립변수와 종속변수를 병합(dictionary -> matrix-data)
iris_df= pd.DataFrame(X, columns=iris.feature_names)
iris_df['species']= iris.target
iris_df.head()
```

```python
# 독립변수의 이름 바꾸기
iris_df = iris_df.rename(
    columns={
        "sepal length (cm)": "sepal_length",
        "sepal width (cm)": "sepal_width",
        "petal length (cm)": "petal_length",
        "petal width (cm)": "petal_width",
    })
iris_df.head()
```

```python
y = iris_df['species']
x = iris_df[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
#훈련용 데이터와 평가용 데이터 분할하기 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)

#로지스틱 회귀 분석: (1) 모델 생성 
LR_iris = LogisticRegression()
```

```python
#로지스틱 회귀 분석: (2) 모델 훈련 및 모형적합결과
reg=LR_iris.fit(x_train, y_train)
print(iris.feature_names, reg.coef_)
print(iris.feature_names, np.exp(reg.coef_))
print( "R²=", reg.score(x_train, y_train))
```

```python
#로지스틱 회귀 분석: (3) 훈련자료와 평가자료에 대한 예측 수행 -> 예측 결과 Y_predict 구하기 
y_fit=LR_iris.predict(x_train)
y_pred = LR_iris.predict(x_test)
```

```python
#혼동행렬
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, y_fit)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# 모형평가(분류정확도_보고서) 종합정리
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
